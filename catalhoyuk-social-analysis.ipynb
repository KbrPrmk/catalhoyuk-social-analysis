EĞİTİM FONKSİYONU — GridSearchCV + Stratified K-Fold

```python
def train_and_evaluate(X, y, models_dict, n_splits=5,
                       scoring='f1_macro', problem='multiclass'):
    
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    best_estimators = {}
    cv_score_matrix = {}      # Friedman testi için: her model, her fold skoru

    for name, (model, param_grid) in models_dict.items():
        print(f"\n{'='*50}")
        print(f"  Model: {name}")
        print(f"{'='*50}")

        grid = GridSearchCV(
            model,
            param_grid,
            cv=skf,
            scoring=scoring,
            n_jobs=-1,
            refit=True
        )
        grid.fit(X, y)

        best_estimators[name] = grid.best_estimator_
        print(f"  En iyi parametreler : {grid.best_params_}")
        print(f"  En iyi CV skoru     : {grid.best_score_:.4f}")

        # Her fold'un skorunu ayrı kaydet (Friedman için)
        cv_res = cross_validate(
            grid.best_estimator_, X, y,
            cv=skf,
            scoring=scoring,
            return_train_score=True
        )
        cv_score_matrix[name] = cv_res['test_score']
        print(f"  CV test skorları    : "
              f"{cv_res['test_score'].round(3)}")
        print(f"  Ortalama ± Std      : "
              f"{cv_res['test_score'].mean():.3f} "
              f"± {cv_res['test_score'].std():.3f}")

    return best_estimators, cv_score_matrix
```

FRİEDMAN TESTİ — Model Farklarının İstatistiksel Anlamlılığı

```python
def friedman_test(cv_matrix, soru_label):
    print(f"\n{'='*60}")
    print(f"  FRİEDMAN TESTİ — {soru_label}")
    print(f"{'='*60}")

    scores = list(cv_matrix.values())
    model_names = list(cv_matrix.keys())

    # Friedman testi: en az 3 model gerektirir
    stat, p = friedmanchisquare(*scores)

    print(f"  Friedman χ² = {stat:.4f},  p = {p:.4f}")
    if p < 0.05:
        print("  → p < 0.05: Modeller arasında istatistiksel olarak")
        print("    ANLAMLI bir performans farkı mevcuttur.")
    else:
        print("  → p ≥ 0.05: Modeller arasında istatistiksel olarak")
        print("    ANLAMLI bir fark saptanamamıştır.")

    # Skor tablosu
    score_df = pd.DataFrame(cv_matrix,
                             index=[f'Fold {i+1}' for i in range(5)])
    print("\n  CV Skor Matrisi:")
    print(score_df.round(4).to_string())

    return stat, p
```

```python
def detailed_report(best_estimators, X, y, class_names, soru_label):
    print(f"\n{'='*60}")
    print(f"  SINIFLANDIRMA RAPORU — {soru_label}")
    print(f"{'='*60}")

    X_tr, X_te, y_tr, y_te = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=42
    )

    fig, axes = plt.subplots(1, len(best_estimators),
                              figsize=(6 * len(best_estimators), 5))
    if len(best_estimators) == 1:
        axes = [axes]

    results = {}
    for ax, (name, model) in zip(axes, best_estimators.items()):
        model.fit(X_tr, y_tr)
        y_pred = model.predict(X_te)

        print(f"\n  [{name}]")
        print(classification_report(y_te, y_pred,
                                     target_names=class_names,
                                     zero_division=0))

        cm = confusion_matrix(y_te, y_pred)
        ConfusionMatrixDisplay(cm, display_labels=class_names).plot(
            ax=ax, cmap='Blues', colorbar=False
        )
        ax.set_title(f'{name}\n({soru_label})', fontsize=10)

        results[name] = {
            'f1_macro': f1_score(y_te, y_pred, average='macro',
                                  zero_division=0),
            'model': model
        }

    plt.tight_layout()
    plt.show()
    return results
```

```python
from sklearn.metrics import roc_curve, auc

X_tr2, X_te2, y_tr2, y_te2 = train_test_split(
    X_q2, y_q2, test_size=0.2, stratify=y_q2, random_state=42
)

print(f"Test seti sınıf dağılımı: {pd.Series(y_te2).value_counts().to_dict()}")

fig, ax = plt.subplots(figsize=(8, 6))

for name, model in best_q2.items():
    model.fit(X_tr2, y_tr2)
    
    # Modelin kaç sınıf gördüğünü kontrol et
    if hasattr(model, 'classes_'):
        seen_classes = model.classes_
    elif hasattr(model, 'named_steps'):          # Pipeline (SVM)
        seen_classes = model.named_steps['clf'].classes_
    else:
        seen_classes = np.unique(y_tr2)
    
    print(f"{name} — eğitimde görülen sınıflar: {seen_classes}")
    
    # Her iki sınıf da yoksa ROC çizilemez, atla
    if len(seen_classes) < 2:
        print(f"  → {name}: tek sınıf, ROC çizilemiyor, atlanıyor.")
        continue
    
    # predict_proba: sütun 1 pozitif sınıfın olasılığı
    try:
        y_prob = model.predict_proba(X_te2)[:, 1]
    except Exception as e:
        print(f"  → {name}: predict_proba hatası: {e}")
        continue
    
    # Test setinde iki sınıf da var mı?
    if len(np.unique(y_te2)) < 2:
        print(f"  → Test setinde tek sınıf var, ROC çizilemiyor.")
        continue
    
    fpr, tpr, _ = roc_curve(y_te2, y_prob)
    roc_auc = auc(fpr, tpr)
    ax.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.3f})', linewidth=2)

ax.plot([0, 1], [0, 1], 'k--', label='Rastgele Sınıflandırıcı')
ax.set_title('ROC Eğrisi — Any Gift Tahmini (Soru 2)', fontsize=13)
ax.set_xlabel('Yanlış Pozitif Oranı (1 - Özgüllük)')
ax.set_ylabel('Doğru Pozitif Oranı (Duyarlılık)')
ax.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
